```python
%matplotlib inline
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis,color="cyan",label='Chicago')
ax.invert_yaxis() 
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.set_title('PyBer Fare by Month')
ax.legend('Chicago',loc='best',bbox_to_anchor=(0.5, 1.0))
plt.legend()
#bbox_inches='tight'), borderaxespad=0.)
#,fontsize='xx-small',bbox_to_anchor=(0.5, 1.0)

plt.show()
```

```python
plt.plot(x_axis, y_axis, 'o')
```

```python
plt.scatter(y_axis, x_axis,s=y_axis,c='red')
plt.gca().invert_yaxis()
plt.xlim(0, 50)
plt.xlabel('Date')
plt.ylabel('Fare($)')
plt.title('PyBer Fare by Month')
plt.legend('Chicago',loc='best')

plt.show()
```

```python
#y_axis_larger = []
#for data in y_axis:
#  y_axis_larger.append(data*3)
#plt.scatter(y_axis, x_axis,s=y_axis_larger,c='red')
plt.scatter(y_axis, x_axis, s = [i * 3 for i in y_axis])
plt.gca().invert_yaxis()
plt.xlim(0, 50)
plt.xlabel('Date')
plt.ylabel('Fare($)')
plt.title('PyBer Fare by Month')
plt.legend('Chicago',loc='best')
plt.show
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [i * 5 for i in y_axis],c='skyblue',alpha=0.8, edgecolors='black',linewidth=2)
ax.invert_yaxis()
#ax.xlimit(0,50)
ax.set_xlim(xmin=0.0, xmax=50)
ax.set_xlabel('Fare($)')
ax.set_ylabel('Date')
ax.set_title('PyBer Fare by Month')
ax.legend('Boston',loc='best')
```

```python
plt.subplots(figsize=(8, 8))
plt.pie(y_axis, labels=x_axis)
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, explode=explode_values, labels=x_axis, autopct='%.1f%%')

plt.show()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```

```python
# create the same pie chart using the object-oriented approach.

colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
fig, ax = plt.subplots(figsize=(8, 8))
#plt.subplots(figsize=(8, 8))
ax.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%',
    counterclock=False,
    shadow=True,
    startangle=90)
#ax.pie(y_axis,labels=x_axis)
plt.show()

```
